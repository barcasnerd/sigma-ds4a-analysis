## TEXT PREPROCESSING

This notebook has the function used to clean the text. It applies to the model, and for the input text that is going to be categorized  

```python
import nltk # imports the natural language toolkit
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize     # Librería para tratamiento de "Tokens"
import re
from spellchecker import SpellChecker

spell = SpellChecker(language='es')

stop_words = stopwords.words('spanish')

def stripchars(data):
    data = data.lower()
    data = data+" "
    data = data.replace("á","a").replace("é","e").replace("í","i").replace("ó","o").replace("ú","u")
    p = re.compile(r'[^\w\s\d]')
    p = p.sub(' ', data)
    p = re.sub(" \d+", " ", p)
    return p

def remove_stop_words(review):
    tokens = word_tokenize(review)  
    text_with_no_stop_words = [spell.correction(token) for token in tokens if not token in stop_words]
    reformed_sentence = ' '.join(text_with_no_stop_words)
    return reformed_sentence
```

```python
txt = "El gps de mi vehiclo me está enviando a la direccon incorecta 834 vces segun la ordn OT8325 y oT_8321"
```

```python
stripchars(txt)
```

```python
remove_stop_words(txt)
```

```python
remove_stop_words(stripchars(txt))
```

```python
txt = "En el perfil 12, desde el formulario de resoluciones externas, no se debe dar un nuevo numero de  resolución si la digitan en el campo de -numero de resolución-, y de igual manera con la fecha de la resolución"
remove_stop_words(stripchars(txt))
```
